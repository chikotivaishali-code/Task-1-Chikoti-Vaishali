```python
from IPython.display import display, clear_output
import time
from IPython.display import display, clear_output
import time
bot_responses = {
    "hello": "Hi there! Welcome to DecodeLabs Bot.",
    "hi": "Hello! How can I assist?",
    "bye": "Goodbye! Have a great day.",
    "exit": "See you later!",
    "quit": "Thanks for chatting!",
    "how are you": "I'm doing great as an AI bot!",
    "name": "I'm the Rule-Based Chatbot from DecodeLabs."
}

print("Bot: Hello! Chat with me (type 'bye', 'exit', or 'quit' to end).")

while True:
    user_message = input("You: ").lower().strip()
    
    if user_message in ["bye", "exit", "quit"]:
        print("Bot: " + bot_responses.get(user_message, "Goodbye!"))
        break
    
    bot_reply = bot_responses.get(user_message, "Sorry, I do not understand. Try 'hello' etc.")
    print(f"Bot: {bot_reply}")
    time.sleep(0.5)
```
